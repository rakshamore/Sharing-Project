```python
def plotdef(df,col,title,hue =None):
    
    sns.set_style("darkgrid", {"grid.color": ".1", "grid.linestyle": ":"})
    sns.set_context("paper")
    plt.rcParams["axes.labelsize"] = 10
    plt.rcParams['axes.titlesize'] = 12
    plt.rcParams['axes.titlepad'] = 20
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) 
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette = ["blue","r"]) 
        
    plt.show()
```

```python
plotdef(payother,col='AMT_INCOME_RANGE',title='Income Range',hue='CODE_GENDER')
```

```python
prev_df = prev_df[~(np.isnan(prev_df['AMT_ANNUITY']))]
```

```python
prev_df.isnull().sum()/len(prev_df.index)
```

```python
cloutlier = ['AMT_ANNUITY','AMT_APPLICATION','AMT_CREDIT','AMT_GOODS_PRICE','DAYS_DECISION','CNT_PAYMENT']
for col in cloutlier: 
    q1 = prev_df[col].quantile(0.25)
    q3 = prev_df[col].quantile(0.75)
    iqr = q3-q1
    range_low  = q1-1.5*iqr
    range_high = q3+1.5*iqr
    prev_df = prev_df.loc[(prev_df[col] > range_low) & (prev_df[col] < range_high)]

prev_df.shape
```

```python

col_list = ['NAME_CONTRACT_TYPE','NAME_CONTRACT_STATUS','NAME_CLIENT_TYPE','NAME_PORTFOLIO','CHANNEL_TYPE']            
k=0
plt.figure(figsize=(20,22))
for col in col_list:    
    k=k+1
    plt.subplot(3, 3,k)    
    ax = sns.countplot(x = col , data = prev_df )
    temp = ax.set_xticklabels(ax.get_xticklabels(), rotation = 45, horizontalalignment='right')
```
